```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
numerical_df = pd.read_csv('numerical_df.csv')
numerical_df
```

```python
X = pd.concat([numerical_df,categorical_df], axis=1)
X
```

```python
print(X.columns.tolist())
```

```python
y=X['total claim amount']
X=X.drop('total claim amount', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
print('X_train shape is:',X_train.shape)
print('y_train shape is:', y_train.shape)
print('X_test shape is:', X_test.shape)
print('y_test shape is:', y_test.shape)
```

```python
X_train["coverage"] = X_train["coverage"].map({"Basic" : 0, "Extended" :.5, "Premium" : 1})
```

```python
X_test["coverage"] = X_test["coverage"].map({"Basic" : 0, "Extended" :.5, "Premium" : 1})
```

```python
X_train["employmentstatus"] = X_train["employmentstatus"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":1})
```

```python
X_test["employmentstatus"] = X_test["employmentstatus"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":1})
```

```python
X_train["location code"] =X_train["location code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
```

```python
X_test["location code"] =X_test["location code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
```

```python
X_train["vehicle size"] =X_train["vehicle size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
```

```python
X_test["vehicle size"] =X_test["vehicle size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
```

```python
X_train["education"] = X_train["education"].map({"High School or Below" : 0, "College" :.33, "Bachelor" : .66,"Post-Uni" : 1})
```

```python
X_test["education"] = X_test["education"].map({"High School or Below" : 0, "College" :.33, "Bachelor" : .66,"Post-Uni" : 1})
```

```python
X_train_1_hot_col = X_train[['state','response','gender','marital status','policy type','renew offer type','sales channel','vehicle class']]
X_train_1_hot_col
```

```python
X_test_1_hot_col = X_test[['state','response','gender','marital status','policy type','renew offer type','sales channel','vehicle class']]
X_test_1_hot_col
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_1_hot_col)

cols = encoder.get_feature_names_out(input_features = X_train_1_hot_col.columns)

X_train_col_encoded = pd.DataFrame(encoder.transform(X_train_1_hot_col).toarray(),columns=cols)
X_test_col_encoded = pd.DataFrame(encoder.transform(X_test_1_hot_col).toarray(),columns=cols)

X_train_col_encoded
```

```python
X_train = X_train.drop(['state','response','education','gender','marital status','policy type','renew offer type','sales channel','vehicle class'],axis = 1)
X_train=X_train.reset_index()
X_train
```

```python
X_test = X_test.drop(['state','response','education','gender','marital status','policy type','renew offer type','sales channel','vehicle class'],axis = 1)
X_test = X_test.reset_index()
X_test
```

```python
X_train = X_train.drop('index', axis=1)
X_test = X_test.drop('index',axis=1)
```

```python
X_train_encoded= pd.concat([X_train,X_train_col_encoded],axis=1)
X_train_encoded= X_train_encoded.reset_index()
X_train_encoded= X_train_encoded.drop('index',axis=1)
X_train_encoded
```

```python
X_test_encoded= pd.concat([X_test,X_test_col_encoded],axis=1)
X_test_encoded= X_test_encoded.reset_index()
X_test_encoded= X_test_encoded.drop('index',axis=1)
X_test_encoded
```

```python
X_train_to_scale = X_train[['customer lifetime value','income','monthly premium auto','months since last claim','months since policy inception','number of open complaints','number of policies']]
X_train_to_scale
X_test_to_scale = X_test[['customer lifetime value','income','monthly premium auto','months since last claim','months since policy inception','number of open complaints','number of policies']]
X_test_to_scale
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
transformer = MinMaxScaler().fit(X_train_to_scale)
```

```python
X_norm1 = transformer.transform(X_train_to_scale)
X_train_scaled = pd.DataFrame(X_norm1, columns=X_train_to_scale.columns)
X_train_scaled = X_train_scaled.reset_index()
```

```python
X_norm2 = transformer.transform(X_test_to_scale)
X_test_scaled = pd.DataFrame(X_norm2, columns=X_test_to_scale.columns)
X_test_scaled= X_test_scaled.reset_index()
```

```python
X_train = X_train.drop(['customer lifetime value','income','monthly premium auto','months since last claim','months since policy inception','number of open complaints','number of policies'],axis = 1)
X_test = X_test.drop(['customer lifetime value','income','monthly premium auto','months since last claim','months since policy inception','number of open complaints','number of policies'], axis = 1)
```

```python
X_train_s_e = pd.concat([X_train, X_train_scaled],axis = 1)
X_test_s_e = pd.concat([X_test, X_test_scaled], axis = 1)
```

```python
X_train_s_e.isna().sum()
```

```python
def models_automation(models, X_tr, y_tr,X_te, y_te):
    for model in models:
        model.fit(X_tr, y_tr)
        print(f"{model.__class__.__name__}: Train -> {model.score(X_tr, y_tr)}, Test -> {model.score(X_te, y_te)}")
```

```python
model_list = [LinearRegression(),SGDRegressor(),KNeighborsRegressor(), MLPRegressor(),DecisionTreeRegressor(),RandomForestRegressor()]
models_automation(model_list, X_train_s_e, y_train,X_test_s_e, y_test)
```
